```python
# Name : Heris Tugimin
# ==============================================================================
# Step 0: Setup and Installation
# ==============================================================================

!pip install --quiet autogluon

# Import necessary libraries
import pandas as pd
import numpy as np
from autogluon.tabular import TabularPredictor
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully.")

# ==============================================================================
# Step 1: Load Data
# ==============================================================================

try:
    train_df = pd.read_csv('train.csv', parse_dates=['datetime'])
    test_df = pd.read_csv('test.csv', parse_dates=['datetime'])
    submission_df = pd.read_csv('sampleSubmission.csv', parse_dates=['datetime'])
except FileNotFoundError:
    print("="*80)
    print("ERROR: 'train.csv' or 'test.csv' not found.")
    print("Please make sure these files are in the same directory as your notebook.")
    print("You can download them from the Kaggle competition page:")
    print("https://www.kaggle.com/competitions/bike-sharing-demand/data")
    print("="*80)
    # Exit the script if files are not found
    exit()
    
# Keep an original copy of the test dataframe for creating the final submission file.
test_df_original = test_df.copy()

print("\nData loaded successfully.")
print("\nTraining data sample:")
print(train_df.head())

# ==============================================================================
# Step 2: Handle the RMSLE Evaluation Metric
# ==============================================================================

train_df = train_df.drop(columns=['casual', 'registered'])

# Apply the log transformation to the target variable 'count'
train_df['count'] = np.log1p(train_df['count'])

print("\nTarget variable 'count' has been log-transformed to optimize for RMSLE.")

# ==============================================================================
# Run 1: Baseline Model 
# ==============================================================================
print("\n--- Starting Run 1: Baseline Model ---")

# Initialize TabularPredictor.
# We set `eval_metric='root_mean_squared_error'`. Since our target is log-transformed,
# minimizing RMSE on this target is equivalent to minimizing RMSLE on the original target.
baseline_predictor = TabularPredictor(
    label='count',
    eval_metric='root_mean_squared_error',
    path='./ag_models/baseline' # Specifies a directory to save models
)

# Fit the model using a medium quality preset for a good balance of speed and accuracy.
baseline_predictor.fit(
    train_data=train_df,
    time_limit=360, # Time limit in seconds
    presets='medium_quality'
)

print("\nBaseline model training complete.")
baseline_predictor.leaderboard()

# Generate predictions on the test data.
baseline_predictions_log = baseline_predictor.predict(test_df)

# Inverse transform the predictions to get them back to the original scale.
# np.expm1 calculates exp(x) - 1.
baseline_predictions = np.expm1(baseline_predictions_log)

# Ensure predictions are non-negative.
baseline_predictions[baseline_predictions < 0] = 0

# Create the submission file.
submission_baseline = submission_df.copy()
submission_baseline['count'] = baseline_predictions
submission_baseline.to_csv('submission_baseline.csv', index=False)

print("\n'submission_baseline.csv' created successfully.")

# ==============================================================================
# Step 3: Exploratory Data Analysis & Feature Engineering
# ==============================================================================
print("\n--- Starting Run 2: Model with Feature Engineering ---")

def create_features(df):
    """Creates time-series features from the datetime column."""
    df['hour'] = df['datetime'].dt.hour
    df['day_of_week'] = df['datetime'].dt.dayofweek
    df['month'] = df['datetime'].dt.month
    df['year'] = df['datetime'].dt.year
    # It might be useful to have a feature for the day of the year as well
    df['day_of_year'] = df['datetime'].dt.dayofyear
    return df

# Create new dataframes with the engineered features
train_df_featured = create_features(train_df.copy())
test_df_featured = create_features(test_df.copy())

print("\nFeature engineering complete. New features: 'hour', 'day_of_week', 'month', 'year', 'day_of_year'")
print(train_df_featured.head())

# ==============================================================================
# Run 2: Retrain Model with New Features
# ==============================================================================
featured_predictor = TabularPredictor(
    label='count',
    eval_metric='root_mean_squared_error',
    path='./ag_models/featured'
)

featured_predictor.fit(
    train_data=train_df_featured,
    time_limit=360,
    presets='medium_quality'
)

print("\nModel training with new features complete.")
featured_predictor.leaderboard()

# Generate and process predictions
featured_predictions_log = featured_predictor.predict(test_df_featured)
featured_predictions = np.expm1(featured_predictions_log)
featured_predictions[featured_predictions < 0] = 0

# Create submission file
submission_featured = submission_df.copy()
submission_featured['count'] = featured_predictions
submission_featured.to_csv('submission_featured.csv', index=False)

print("\n'submission_featured.csv' created successfully.")

# ==============================================================================
# Run 3: Retrain with Hyperparameter Optimization (HPO)
# ==============================================================================
print("\n--- Starting Run 3: Model with Hyperparameter Optimization ---")

# This run uses the 'best_quality' preset which enables more extensive HPO,
# longer training times for individual models, and more complex ensembling.
hpo_predictor = TabularPredictor(
    label='count',
    eval_metric='root_mean_squared_error',
    path='./ag_models/hpo'
)

hpo_predictor.fit(
    train_data=train_df_featured,
    time_limit=600, # Increased time limit for better results
    presets='best_quality'
)

print("\nModel training with HPO complete.")
hpo_predictor.leaderboard()

# Generate and process predictions
hpo_predictions_log = hpo_predictor.predict(test_df_featured)
hpo_predictions = np.expm1(hpo_predictions_log)
hpo_predictions[hpo_predictions < 0] = 0

# Create submission file
submission_hpo = submission_df.copy()
submission_hpo['count'] = hpo_predictions
submission_hpo.to_csv('submission_hpo.csv', index=False)

print("\n'submission_hpo.csv' created successfully.")

# ==============================================================================
# Step 4: Final Report Generation
# ==============================================================================
print("\n--- Generating Final Report Data ---")

# These are placeholder values.
# IMPORTANT: Replace these with your actual scores from Kaggle!
KAGGLE_SCORE_BASELINE = 1.93846
KAGGLE_SCORE_FEATURED = 0.42021
KAGGLE_SCORE_HPO = 0.55616

# Get internal evaluation scores from the predictors
eval_metric_baseline = abs(baseline_predictor.info()['best_model_score_val'])
eval_metric_featured = abs(featured_predictor.info()['best_model_score_val'])
eval_metric_hpo = abs(hpo_predictor.info()['best_model_score_val'])

# Create a summary DataFrame
summary_data = {
    'Run': ['1. Baseline', '2. With Features', '3. With HPO'],
    'Model Configuration': [
        'Default AutoGluon `medium_quality` preset on raw data.',
        'Added time-based features (`hour`, `day_of_week`, etc.).',
        'Used `best_quality` preset for extensive HPO and ensembling.'
    ],
    'Internal RMSE (on log-data)': [
        f"{eval_metric_baseline:.5f}",
        f"{eval_metric_featured:.5f}",
        f"{eval_metric_hpo:.5f}"
    ],
    'Kaggle RMSLE Score': [
        KAGGLE_SCORE_BASELINE,
        KAGGLE_SCORE_FEATURED,
        KAGGLE_SCORE_HPO
    ]
}
summary_df = pd.DataFrame(summary_data)

print("\n--- Project Summary Table ---")
# Using to_markdown for a clean, readable output
print(summary_df.to_markdown(index=False))

# --- Create Graphs ---

# Graph 1: Internal Evaluation Metric
plt.figure(figsize=(12, 6))
sns.barplot(x='Run', y='Internal RMSE (on log-data)', data=summary_df, palette='viridis')
plt.title('Internal Model Performance (RMSE on Log-Transformed Data)', fontsize=16)
plt.ylabel('Validation Score (Lower is Better)', fontsize=12)
plt.xlabel('Model Version', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Graph 2: Kaggle Scores
plt.figure(figsize=(12, 6))
sns.barplot(x='Run', y='Kaggle RMSLE Score', data=summary_df, palette='plasma')
plt.title('Kaggle Leaderboard Score (RMSLE)', fontsize=16)
plt.ylabel('Kaggle Score (Lower is Better)', fontsize=12)
plt.xlabel('Model Version', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

print("\n--- End of Project ---")
```
